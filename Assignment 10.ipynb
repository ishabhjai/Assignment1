```r
# Load the air quality dataset
data(airquality)

# Assign column names to the dataset
colnames(airquality) <- c("Ozone", "Solar.R", "Wind", "Temp", "Month", "Day")
```

```r
# a histogram with custom colors
hist(airquality$Ozone, col = "darkblue", border = "white")
```

```r
# a histogram with no axis and custom labels
hist(airquality$Ozone, col = "darkblue", border = "white", axes = FALSE, xlab = "Ozone Levels", ylab = "Frequency")
```

```r
# a histogram with custom axis limits
hist(airquality$Ozone, col = "darkblue", border = "white", xlim = c(0, 100), ylim = c(0, 30))
```

```r
# Create a histogram with density and add a density curve
hist(airquality$Ozone, col = "darkblue", border = "white", prob = TRUE)
lines(density(airquality$Ozone), col = "red", lwd = 2)
```
