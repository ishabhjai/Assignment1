```r
# Create a list of marks for three subjects for 10 students
marks_list <- list(
  Math = c(80, 75, 90, 85, 92, 78, 88, 83, 79, 87),
  English = c(70, 65, 80, 75, 82, 68, 78, 73, 69, 77),
  `Data Interpretation` = c(85, 80, 95, 90, 92, 88, 90, 87, 83, 91)
)

# Convert the list into a data frame
marks_df <- as.data.frame(marks_list)

# Calculate total marks, average, maximum, and minimum marks for each subject
subject_stats <- data.frame(
  Subject = character(),
  TotalMarks = numeric(),
  AverageMarks = numeric(),
  MaxMarks = numeric(),
  MinMarks = numeric(),
  stringsAsFactors = FALSE
)

for (subject in names(marks_df)) {
  total_marks <- sum(marks_df[[subject]])
  average_marks <- mean(marks_df[[subject]])
  max_marks <- max(marks_df[[subject]])
  min_marks <- min(marks_df[[subject]])
  
  subject_stats <- rbind(subject_stats, data.frame(
    Subject = subject,
    TotalMarks = total_marks,
    AverageMarks = average_marks,
    MaxMarks = max_marks,
    MinMarks = min_marks
  ))
}

# Output the subject statistics
print(subject_stats)
```
