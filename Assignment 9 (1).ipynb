```r
# Create a list with different types of elements
my_list <- list("Hello", 123, c(1, 2, 3), TRUE)

# Access the first element in the list
first_element <- my_list[[1]]
cat("First element:", first_element, "\n\n")

# Give names to the elements in the list
names(my_list) <- c("String", "Number", "Vector", "Logical")

# Add an element at a specific position in the list
new_element <- "New"
my_list <- append(my_list, new_element, after = 2)

# Remove an element from the list
my_list <- my_list[-3]

# Print the fourth element
fourth_element <- my_list[[4]]
cat("Fourth element:", fourth_element, "\n\n")

# Update the third element
my_list[[3]] <- c(4, 5, 6)

# Print the updated list
print(my_list)
```
