```r
# Input
n <- as.integer(readline("Enter a positive integer: "))

# Initialize variables
sum <- 0
i <- 1

# Loop until i reaches n
while (i <= n) {
  # Add i to the sum
  sum <- sum + i
  
  # Increment i
  i <- i + 1
}

# Output the sum
cat("The sum of natural numbers up to", n, "is", sum, "\n")
```
