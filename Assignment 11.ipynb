```r
employee_data <- read.csv("input.csv")
```

```r
num_rows <- nrow(employee_data)
num_cols <- ncol(employee_data)
```

```r
max_salary <- max(employee_data$salary)
```

```r
employee_max_salary <- subset(employee_data, salary == max_salary)
```

```r
employees_IT_dept <- subset(employee_data, dept == "IT")
```

```r
employees_high_salary <- subset(employees_IT_dept, salary > 20000)
write.csv(employees_high_salary, "output.csv", row.names = FALSE)
```

```r
cat("Total number of rows:", num_rows, "\n")
cat("Total number of columns:", num_cols, "\n")
cat("Maximum salary:", max_salary, "\n")
cat("Employee with maximum salary:\n")
print(employee_max_salary)
cat("Employees in the IT Department:\n")
print(employees_IT_dept)
```
