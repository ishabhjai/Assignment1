```r
# Function to perform addition
add <- function(a, b) {
  return(a + b)
}

# Function to perform subtraction
subtract <- function(a, b) {
  return(a - b)
}

# Function to perform multiplication
multiply <- function(a, b) {
  return(a * b)
}

# Function to perform division
divide <- function(a, b) {
  if (b == 0) {
    return("Error: Division by zero is not allowed.")
  } else {
    return(a / b)
  }
}

# Main calculator program
operation <- as.character(readline("Enter the operation (+, -, *, /): "))

# Input two numbers
num1 <- as.numeric(readline("Enter the first number: "))
num2 <- as.numeric(readline("Enter the second number: "))

# Perform the operation 
result <- add(num1, num2)
result1 <- subtract(num1, num2)
result2 <- multiply(num1, num2)
result3 <- divide(num1, num2)

# Output the result
cat("Result:", result, "\n")
cat("Result1:", result1, "\n")
cat("Result2:", result2, "\n")
cat("Result3:", result3, "\n")
```
