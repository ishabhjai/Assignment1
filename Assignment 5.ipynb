```r
# Create a list of numbers from 1 to 50
number_list <- 1:50

# Input the number to search
search_number <- as.integer(readline("Enter a number to search: "))

# Perform the search
if (search_number %in% number_list) {
  cat("Search is successful. The number", search_number, "is in the list.\n")
} else {
  cat("The number", search_number, "is not in the list.\n")
}
```
