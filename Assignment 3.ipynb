```r
# Input
marks <- as.integer(readline("Enter the marks obtained: "))

# Grading system
if (marks >= 800 && marks <= 1000) {
  grade <- "A+"
} else if (marks >= 700 && marks < 800) {
  grade <- "A"
} else if (marks >= 500 && marks < 700) {
  grade <- "B+"
} else if (marks >= 400 && marks < 500) {
  grade <- "B"
} else if (marks >= 150 && marks < 400) {
  grade <- "C"
} else {
  grade <- "D"
}

# Output
cat("Grade:", grade, "\n")
```
